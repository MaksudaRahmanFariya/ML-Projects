```python
import scipy.optimize as opt
import statsmodels.api as sm
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.mlab as mlab
```

```python
dt_df = pd.read_csv("/kaggle/input/framingham/framingham.csv")
```

```python
dt_df.drop(['education'],inplace=True,axis=1)    
#The inplace parameter controls whether the changes are applied directly to the original DataFrame or if a new modified copy is returned.
```

```python
dt_df.rename(columns ={'male':'Sex_male'}, inplace = True)
```

```python
dt_df.dropna(axis = 0, inplace = True)
print(dt_df.head(), dt_df.shape)
print(dt_df.TenYearCHD.value_counts())
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='TenYearCHD', data=dt_df,
             palette="BuGn_r")
plt.show()
```

```python
laste = dt_df['TenYearCHD'].plot()
plt.show(laste)
```

```python
X = np.asarray(dt_df[['age', 'Sex_male', 'cigsPerDay', 
                           'totChol', 'sysBP', 'glucose']])
y = np.asarray(dt_df['TenYearCHD'])

# normalization of the dataset
X = preprocessing.StandardScaler().fit(X).transform(X)

# Train-and-Test -Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( 
        X, y, test_size = 0.3, random_state = 4)

print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy of the model is =', 
      accuracy_score(y_test, y_pred))


```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test,y_pred)
cof_matrix = pd.DataFrame(data = cm, columns = ['Predicted:0', 'Predicted:1'],
                         index =['Actual:0', 'Actual:1'])
plt.figure(figsize=(8,5))
sns.heatmap(cof_matrix, annot = True, fmt = 'd', cmap="Greens")
plt.show()
print('The details for confusion matrix is =')
print (classification_report(y_test, y_pred))
```
